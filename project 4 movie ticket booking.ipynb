```python
# Movie Ticket Booking System
```

```python

def show_movies():
    print("Available Movies:")
    print("1. Jawan")
    print("2. Avatar 2")
    print("3. Pushpa 2")

def select_movie(choice):
    if choice == 1:
        return "Jawan"
    elif choice == 2:
        return "Avatar 2"
    elif choice == 3:
        return "Pushpa 2"
    else:
        return None

def calculate_amount(tickets, price=200):
    return tickets * price

def select_seats(tickets):
    seats = []
    print("\nChoose your seat numbers:")
    for i in range(tickets):
        seat = input(f"Enter seat number {i+1}: ")
        seats.append(seat)
    return seats


print("===== MOVIE TICKET BOOKING =====")

show_movies()
choice = int(input("Choose movie (1-3): "))
movie = select_movie(choice)

if movie is None:
    print("Invalid movie choice!")
    exit()

name = input("Enter your name: ")
tickets = int(input("How many tickets? "))


selected_seats = select_seats(tickets)


total_amount = calculate_amount(tickets)


print("\n===== BOOKING CONFIRMED =====")
print("Name:", name)
print("Movie:", movie)
print("Total Tickets:", tickets)
print("Seats:", ", ".join(selected_seats))
print("Total Amount:", total_amount)
print("Enjoy your movie! ")
```
