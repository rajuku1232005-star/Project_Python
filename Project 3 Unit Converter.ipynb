```python
print("===== UNIT CONVERTER =====")

print("""
1. Km → Miles
2. Miles → Km
3. Celsius → Fahrenheit
4. Fahrenheit → Celsius
5. Meter → Centimeter
6. Centimeter → Meter
7. Kg → Gram
8. Gram → Kg
9. Hours → Minutes
10. Minutes → Hours
11. Seconds → Minutes
12. Minutes → Seconds
""")

choice = int(input("Choose option (1-12): "))

if choice == 1:
    km = float(input("Enter km: "))
    print("Miles =", km * 0.621371)

elif choice == 2:
    miles = float(input("Enter miles: "))
    print("Km =", miles / 0.621371)

elif choice == 3:
    c = float(input("Enter Celsius: "))
    print("Fahrenheit =", (c * 9/5) + 32)

elif choice == 4:
    f = float(input("Enter Fahrenheit: "))
    print("Celsius =", (f - 32) * 5/9)

elif choice == 5:
    m = float(input("Enter meter: "))
    print("Centimeter =", m * 100)

elif choice == 6:
    cm = float(input("Enter centimeter: "))
    print("Meter =", cm / 100)

elif choice == 7:
    kg = float(input("Enter kg: "))
    print("Gram =", kg * 1000)

elif choice == 8:
    g = float(input("Enter gram: "))
    print("Kg =", g / 1000)

elif choice == 9:
    h = float(input("Enter hours: "))
    print("Minutes =", h * 60)

elif choice == 10:
    mins = float(input("Enter minutes: "))
    print("Hours =", mins / 60)

elif choice == 11:
    sec = float(input("Enter seconds: "))
    print("Minutes =", sec / 60)

elif choice == 12:
    mins = float(input("Enter minutes: "))
    print("Seconds =", mins * 60)

else:
    print("Invalid choice")
```
